```python
from PIL import Image
import os, glob, sys, numpy as np
from keras.utils import np_utils

train_dir = 'C:/Users/황동호/SCHOOL_/3-2/3_2_final_project/archive/chest_xray/train'
test_dir = 'C:/Users/황동호/SCHOOL_/3-2/3_2_final_project/archive/chest_xray/test'
categories = ['NORMAL', 'PNEUMONIA']
np_classes = len(categories)

image_w = 150
image_h = 150


pixel = image_h * image_w * 3

X = []
y = []

for idx, cat in enumerate(categories):
    train_dir_detail = train_dir + "/" + cat
    train_files = glob.glob(train_dir_detail+"/*.jpeg")
    
    test_dir_detail = test_dir + '/' + cat
    test_files = glob.glob(test_dir_detail + '/*.jpeg')

    for i, f in enumerate(train_files):
        try:
            img = Image.open(f)
            img = img.convert("RGB")
            img = img.resize((image_w, image_h))
            data = np.asarray(img)
            #Y는 0 아니면 1이니까 idx값으로 넣는다.
            X.append(data)
            y.append(idx)
            
        except:
            print(cat, str(i)+" 번째에서 에러 ")
            
    for i, f in enumerate(test_files):
        try:
            img = Image.open(f)
            img = img.convert("RGB")
            img = img.resize((image_w, image_h))
            data = np.asarray(img)
            #Y는 0 아니면 1이니까 idx값으로 넣는다.
            X_t.append(data)
            Y_t.append(idx)
            
        except:
            print(cat, str(i)+" 번째에서 에러 ")
            
X_train = np.array(X)
Y_train = np.array(y)
X_test = np.array(X_t)
Y_test = np.array(Y_t)

#xy = (X_train, Y_train)
#np.save("./3_2_final_project/archive/binary_image_data.npy", xy)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import keras.backend.tensorflow_backend as K

#X_train, X_test, y_train, y_test = np.load('./GOOOOOOOOOOOOOOOD/numpy_data/binary_image_data.npy')
print(X_train.shape)
print(X_train.shape[0])
print(X_test.shape)
print(np.bincount(Y_train))
print(np.bincount(Y_test))
#print(np.bincount(y_test))
```

```python
image_w = 150
image_h = 150
X_train = X_train.astype('float32') / 255
#X_test = X_test.astype('float32') / 255


with K.tf_ops.device('/device:GPU:0'):
    model = Sequential()
    model.add(Conv2D(32, (3,3), padding="same", input_shape=X_train.shape[1:], activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(64, (3,3), padding="same", activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(64, (3,3), padding="same", activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(128, (3,3), padding="same", activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(128, (3,3), padding="same", activation="relu"))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Flatten())
    model.add(Dense(256, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    model_dir = './model'
    if not os.path.exists(model_dir):
        os.mkdir(model_dir)
    model_path = model_dir + "/IMAGE_classify.model"
    
    checkpoint = ModelCheckpoint(filepath=model_path, monitor='val_loss', verbose=1, save_best_only=True)
    early_stopping = EarlyStopping(monitor='val_loss', patience=7)
```

```python
model.summary()
```

```python
history = model.fit(X_train, Y_train, batch_size=64, epochs=10, validation_split=0.15,
                    callbacks=[checkpoint, early_stopping])
```

```python
print("정확도 : %.2f " %(model.evaluate(X_test, y_test)[1]))
```
